**Примечание**

Не все файлы подходящие по названию, будут подходить по содержанию. Там может оказаться лишний столбец, например. Ориентируйтесь на столбцы из чека выше - это то, что вас интересует. Остальные столбцы можно просто отбросить.

**Важно**: разделителем файла на выходе должна быть запятая.

```python
def process_files(src_folder, dest_folder):
    import pandas as pd
    import glob
    import re
    import os


    pattern = r'\d{4}-\d{2}-\d{2}-\d{2}-\d{2}-\d+\.csv'


    all_files = glob.glob(os.path.join(src_folder, '*.csv'))


    valid_files = [f for f in all_files if re.match(pattern, os.path.basename(f))]

    if not valid_files:
        print("Нет подходящих файлов для объединения.")
        return

    combined_df = pd.DataFrame()
    required_columns = ['date', 'product', 'store', 'cost']

    for file in valid_files:
        try:
            df = pd.read_csv(file, sep=";")


            df = df.loc[:, df.columns.intersection(required_columns)]


            for col in required_columns:
                if col not in df.columns:
                    df[col] = pd.NA


            df = df[required_columns]

            combined_df = pd.concat([combined_df, df], ignore_index=True)

        except Exception as e:
            print(f"Файл {file} пропущен из-за ошибки: {e}")


    combined_df = combined_df.sort_values(by=['date', 'product'])
    combined_df.reset_index(drop=True, inplace=True)


    for col in combined_df.columns:
        combined_df[col] = combined_df[col].astype(str)


    os.makedirs(dest_folder, exist_ok=True)


    combined_path = os.path.join(dest_folder, 'combined_data.csv')
    combined_df.to_csv(combined_path, index=False)

    print(f"Объединённый файл сохранён: {combined_path}")


src_folder = 'reports-main'
dest_folder = 'comb_reports'
process_files(src_folder, dest_folder)
```
